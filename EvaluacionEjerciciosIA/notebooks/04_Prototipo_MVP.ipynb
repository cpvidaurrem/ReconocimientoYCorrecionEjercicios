```python
import cv2
import torch
from ultralytics import YOLO
import numpy as np
import streamlit as st
from PIL import Image
import time
```

```python
# Carga el modelo YOLOv8s-pose
model_pose = YOLO(r'C:\Users\CHRISTIAN\Documents\EvaluacionEjerciciosIA\models\yolov8s-pose.pt')

# Simula un LSTM preentrenado (placeholder; lo finetunearemos después con datos reales)
class SimpleLSTM:
    def __init__(self):
        self.threshold = 0.7  # Umbral para considerar un ejercicio "correcto"
    def evaluate(self, keypoints):
        # Simulación: calcula una puntuación basada en keypoints (placeholder)
        alignment = np.random.uniform(0.6, 0.9)  # Simula alineación
        depth = np.random.uniform(0.5, 0.8)      # Simula profundidad
        fluency = np.random.uniform(0.7, 0.95)    # Simula fluidez
        score = 0.4 * 0.95 + 0.3 * alignment + 0.3 * depth  # Puntuación ponderada
        return {"exercise": "sentadilla", "score": score, "alignment": alignment, "depth": depth, "fluency": fluency}

lstm = SimpleLSTM()
```

Función para procesar video y keypoints:

```python
def process_frame(frame):
    # Detección de pose con YOLOv8s-pose
    results = model_pose(frame)
    keypoints = results[0].keypoints.xy.cpu().numpy()  # 17 keypoints por persona
    if len(keypoints) > 0:
        keypoints = keypoints[0]  # Tomamos la primera persona detectada
        evaluation = lstm.evaluate(keypoints)
        return keypoints, evaluation
    return None, None

# Captura video
cap = cv2.VideoCapture(0)
st.title("Prototipo - Evaluación de Ejercicios")
placeholder = st.empty()
```

Bucle principal con Streamlit:

```python
while True:
    ret, frame = cap.read()
    if not ret:
        st.write("No se puede acceder a la webcam.")
        break

    keypoints, evaluation = process_frame(frame)
    if evaluation:
        # Muestra resultados
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img_pil = Image.fromarray(frame_rgb)
        with placeholder.container():
            st.image(img_pil, caption="Video en tiempo real", use_column_width=True)
            st.write(f"Ejercicio: {evaluation['exercise']}")
            st.write(f"Puntuación: {evaluation['score']*100:.1f}/100")
            st.write(f"Alineación: {evaluation['alignment']*100:.1f}%")
            st.write(f"Profundidad: {evaluation['depth']*100:.1f}%")
            st.write(f"Fluidez: {evaluation['fluency']*100:.1f}%")
            if evaluation['score'] < lstm.threshold:
                st.error("Alerta: Técnica incorrecta, ajusta tu postura!")

    time.sleep(0.03)  # ~30 fps

cap.release()
```
