```python
import torch

print("üîç Verificando instalaci√≥n de PyTorch...")

# Ver versi√≥n de PyTorch
print(f"Versi√≥n de PyTorch: {torch.__version__}")

# Verificar si CUDA est√° disponible
cuda_available = torch.cuda.is_available()
print(f"¬øCUDA disponible? {cuda_available}")

if cuda_available:
    # Obtener nombre de la GPU
    gpu_name = torch.cuda.get_device_name(0)
    print(f"GPU detectada: {gpu_name}")

    # Crear un tensor y moverlo a la GPU
    try:
        x = torch.rand(3, 3).to('cuda')
        print("‚úÖ Tensor creado exitosamente en la GPU:")
        print(x)
    except Exception as e:
        print("‚ùå Error al crear tensor en GPU:")
        print(e)
else:
    print("‚ö†Ô∏è No se detect√≥ CUDA. Aseg√∫rate de tener los drivers y PyTorch con soporte CUDA instalados.")
```
